### Logistic Regression

```python
# Def HyperParams
penalty=['l2']
tol = [0.001, 0.0001, 0.00001]
C = [100.0, 10.0, 1.00, 0.1, 0.01, 0.001]
fit_intercept = [True, False]
intercept_scaling = [1.0, 0.75, 0.5, 0.25]
class_weight = ['balanced', None]
solver = ['newton-cg', 'sag', 'lbfgs', 'saga']
max_iter=[14000]

param_distributions = dict(penalty=penalty, tol=tol, C=C, fit_intercept=fit_intercept, intercept_scaling=intercept_scaling,
                  class_weight=class_weight, solver=solver, max_iter=max_iter)
```

Penjelasan
1. penalty=['l2'] : ini menentukan jenis regulasi yang akan digunakan oleh model. Regulasi digunakan untuk mencegah overfitting dengan menambahkan penalti pada kompleksitas model. 'l2' adalah regolasi Ridge yang menambahkan penalti proporsional dengan kuadrat dari koefisien.
2. tol: Ini adalah toleransi untuk kriteria konvergensi. Semakin kecil nilai tol, semakin presisi model dalam mencapai konvergensi, namun mungkin memerlukan lebih banyak iterasi.
3. C: adalah params invers dari regularisasi yang lebih kuat, sementara nilai yang lebih besar mengindikasikan regularisasi yang lebih lemah.
4. Fit-intercept: Ini hanya akan digunakan jika fit_intercept diatur ke 'True'. Ini adalah faktor skala untuk intercept. Penggunaannya lebih relevan dalam model solvers seperti liblinear.
5. Class_weight: Menentukan bobot untuk kelas-kelas. Jika diatur ke 'balanced', model akan menyeimbangkan bobot sesuai dengan distribusi kelas dalam data pelatihan.
6. Solver: menentukan algoritma optimasi yang digunakan untuk menentukkan params model terbaik. Berbagai solver memiliki kinerja yang berbeda-beda tergantung pada ukuran dataset dan kompleksitasnya.
7. Max_iter = Ini menentukan jumlah max iterasi yang diizinkan untuk solver konvergen.
8. Param_distribution: Ini adalah kamus (dict) yang menggabungkan semua params di atas. Kamus ini digunakan untuk menentukan ruang pencarian params dalam pencarian hyperparams (misal, RandomizedSearchCV atau GridSearchCV)

#### Experiment 1

```python
logreg = LogisticRegression()
```

```python
grid = GridSearchCV(estimator=logreg, param_grid=param_distributions, scoring='recall', cv=5, n_jobs=-1)
start_time=time.time()
grid_result = grid.fit(x1_train, y1_train)

execute_time = str((time.time() - start_time))
print(f'Best: {grid_result.best_score_} Using  {grid_result.best_params_}')
print(f'Execution Time: {execute_time} s')
```

time.time adalah fungsi dalam modul time di Python yang mengembalikan waktu saat ini dalam bentuk detik sejak 'epoch'. 'epoch/ adalah 00:00:00 UTC pada 1 Jan 1970. Jadi, time.time(), mengembalikan jumlah detik yang digunakan. Penggunaan time.time sering digunakan untuk mengukur durasi waktu suatu proses.

## Kesimpulan:

Dari semua model, rata-rata memiliki accuracy di atas 60% namun memiliki recall di bawah 40%. Artinya, masih banyak nasabah yang sebenarnya berpotensi gagal bayar namun diprediksi tidak akan gagal bayar. Sehingga bisa disampaikan bahwa dalam iterasi pembangunan model kali ini, objektif yang diinginkan masih belum dapat tercapai.

Solusi pengembangan kedepannya yang bisa dilakukan di antaranya:

Memperbanyak sample (jumlah nasabah dgn asumsi dataset yang tersedia saat ini bukan total populasi nasabah)
Melakukan oversampling terhadap kelas minoritas (gagal bayar) agar pembangunan model tidak bias
Memperluas horizon waktu
Mencoba variasi variabel lainnya (menambah variabel baru, atau membuang variabel yang memiliki nilai importance rendah pada hasil terakhir)
Mencoba memperluas kombinasi hyperparameter dalam pembangunan model
Mencoba algoritma supervised machine learning lainnya

# Catatan

max_features adalah params yang menentukan jumlah fitur yang akan dipertimbangkan untuk membagi setiap node dalam pohon keputusan di dalam random forest

cv adalah cross Validation, cv = 5 artinya permodelan dibagi menjadi 5 subset (folds). Model dilatih pada 4 fold dan diuji pada 1 fold, dan ini dilakukan 5 kali dengan setiap fold menjadi data uji satu kali. Hasilnya dirata-rata untuk mengurangi kemungkinan bias dan memberikan eval yang lebih robust terhadap performa model.

gamma: nilai minimal loss reduction yang dibutuhkan pada saat pemecahan cabang. Semakin besar nilai yang ditetapkan, model yang dibangun akan lebih konservatif dan memungkinkan terjadinya underfitting.

Learning_rate: tingkat penyesuaian bobot fitur. Dalam pembangunan model, setiap iterasi menghasilkan bobot untuk fitur-fitur yang dimiliki. Learning_rate berguna untuk menyusutkan nilai bobot tersebut agar model yang dibangun tidak memiliki overfitting.

Scale_pos_weight: pengaturan bobot antar kelas positif (churn) dengan kelas negatif (not churn). Hyperparameter ini sangat berguna ketika dataset yang digunakan merupakan imbalance datasets. Nilai yang biasa digunakan yaitu jumlah dari majority class dibagi dengan jumlah minority class

completeness_score(): berguna untuk mengukur completeness dari klasterisasi, yaitu seberapa baik semua anggota dari 1 class yang benar (true label) diklasterkan ke dalam satu klaster yang sama. Score ini berkisar antara 0.0 (klaster yang buruk/tidak lengkap/tidak bekerja dengan baik) hingga 1.0 (klaster yang sangat lengkap).
